Same number of columns in another dataset and some of them have zero not-Null

````python
# import 3b data
b_df = pd.read_csv("LoanStats3b_securev1.csv",engine="python",error_bad_lines=False,header=1)
````

````python
print(b_df.info(max_cols=200))
````

concatenate two datasets as the full dataset

````python
## concate the 2 datasets into 1
full_df = pd.concat([a_df, b_df])
````

````python
full_df.info()
````

Remove columns which has Nulls in majority of entires (>=90%)

97 columns left after remove columns that below 10% non-null values threshold 

````python
sum_entries = len(full_df)
````

````python
full_clean_df = full_df.dropna(axis=1,thresh=sum_entries*0.1, inplace=False)
````

````python
full_clean_df.info(max_cols=200)
````

````python
# export to csv file
full_clean_df.to_csv("full_clean_data.csv",header="True",index=False)
````
